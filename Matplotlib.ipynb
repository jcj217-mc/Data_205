####**Stack Plot**

Two or more related groups compared

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

#empty lines just to apply color, label, linewidth
plt.plot([],[], color='m', label= 'Sleeping', linewidth = 5)
plt.plot([],[], color='c', label= 'Eating', linewidth = 5)
plt.plot([],[], color='r', label= 'Working', linewidth = 5)
plt.plot([],[], color='k', label= 'Playing', linewidth = 5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m', 'c', 'r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

####**Multiple Plots**

```python
import numpy as np

def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```

####**FIFA Data**

```python
# Load FIFA data
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

### Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')
# google color picker

plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')


plt.show()
```

#### Pie Charts

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
print('Left:', left)

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
print('Right:', right)

# Pie Chart 1
# needs values - so why left and right made above
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left,right], labels = labels, colors=colors, autopct='%.2f %%') 
#.2 is number of decimals, 'f' stand for "float", %% gives % sign on chart

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight
# reset weight category to not include lbs.  
# string right now and need to convert to integer

#reset weight column
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
fifa.Weight[0]
```

```python
plt.style.use('ggplot')

# start counting
light = fifa.loc[fifa.Weight < 125].count()[0]
# the zero gives us only the numbers in that column
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >=175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode= (.4, .1, 0, 0, 0.4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct = '%.2f %%', pctdistance = 0.8, explode =explode)
# pctdistance is distance from center of the circle

plt.show()
```

#### **Box and Whiskers Chart**

How do we compare teams?

```python
plt.style.use('default')

# stretched out figure
plt.figure(figsize=(5,8))

# The overall values for each memeber on team FC Barcelona 
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club =='New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

# will save each individual box so we can style them
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops = {'linewidth':2})
# patch artist needed to do facecolor change
for box in boxes['boxes']:
  # set edge color
  box.set(color='#4286f4', linewidth=2)
  # change Fill color
  box.set(facecolor = '#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')



plt.show()
```
